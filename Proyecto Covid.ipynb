## Limpieza, organización y estandarización de los datos

```python
# Biblioteca
import pandas as pd
import numpy as np
from IPython.display import display
```

```python
# Cargamos la base de datos
try:
  df = pd.read_csv("2019 COVID.csv")
  display(df)
except Exception as e:
  print("verifica la ruta del archivo")
  print(f'Error: {e}')
else:
  print("Lectura finalizada")
```

```python
# Panorama general de los datos.
def info_general_df(df):
  print('Información general del DataFrame')
  print(f'Cantidad de filas y columnas: {df.shape}')
  print("-"*50)
  print(f'Cantidad de datos nulos por columna: \n{df.isnull().sum()}')
  print("-"*50)
  print(f'Cantidad de datos únicos por columna: \n{df.nunique()}')
  print("-"*50)
  print(f'Tipos de datos por columna: \n{df.dtypes}')

info_general_df(df)
```

```python
# Hacemos una copia del dataframe original.
df_current = df.copy()
df_current.head()
```

```python
# Convertimos las columnas en sus tipos de datos correspondientes.
df_current['Date'] = pd.to_datetime(df_current['Date'], errors='coerce')
df_current['Last Update'] = pd.to_datetime(df_current['Last Update'], errors='coerce')

for region in df_current['Region'].unique():
    mean_lat = df_current.loc[df_current['Region'] == region, 'Lat'].mean()
    df_current.loc[(df_current['Region'] == region) & (df_current['Lat'].isna()), 'Lat'] = mean_lat
    mean_long = df_current.loc[df_current['Region'] == region, 'Long'].mean()
    df_current.loc[(df_current['Region'] == region) & (df_current['Long'].isna()), 'Long'] = mean_long

# Imprimir los tipos de datos para asegurarse de que todo se haya convertido correctamente
print("Conversion finalizada")
display(df_current.dtypes)
print("-"*150)
display(df_current)
```

```python
# Procesamiento de datos faltantes.
data_fillna = {
    'Province': "Desconocido", 

    # Generar fechas aleatorias para las columnas de fechas
    'Date': pd.to_datetime(np.random.uniform(df_current['Date'].min().value, df_current['Date'].max().value)).normalize(),
    'Last Update': pd.to_datetime(np.random.uniform(df_current['Last Update'].min().value, df_current['Last Update'].max().value)),

    # Usar la media para completar valores numéricos faltantes
    'Lat': float(df_current['Lat'].mean()),  
    'Long': float(df_current['Long'].mean()), 
}

# Usar fillna con el diccionario
df_current.fillna(data_fillna, inplace=True)
# Verificar la cantidad de nulos restantes
print(df_current.isnull().sum())
print("-"*50)
display(df_current.dtypes)
print("-"*150)
display(df_current)
```

```python
# Obtenemos todas las columnas.
current_name_list=df_current.columns.tolist()
print(current_name_list)

column_name_mapping = {
    'Date': 'date',
    'Province': 'province',
    'Region': 'region',
    'Last Update': 'last update',
    'Confirmed': 'confirmed',
    'Deaths': 'deaths',
    'Recovered': 'recovered',
    'Lat': 'lat',
    'Long': 'long',
}

df_current.rename(columns=column_name_mapping, inplace=True)
```

```python
# Detectar valores duplicados
df_current.drop_duplicates(inplace=True)
info_general_df(df_current)
```

```python
display(df_current)
```

## **Calcular metricas y rangos de datos**

### Defunciones

```python
# Agrupar por region y calcular el numero de muertes 
muertes_por_region = df_current.groupby('region')['deaths'].sum()
muertes_por_region
```

```python
region_max_muertes = muertes_por_region.idxmax()
max_muertes = muertes_por_region.max()
print(f"La región con más muertes es: {region_max_muertes}, con {max_muertes} muertes.")
```

```python
top_10_paises = muertes_por_region.nlargest(10)

print("Los 10 países con más muertes:")
print(top_10_paises)
```

```python
# Aplicamos la funcion de map, para generar un rango del número de muertes 
def deaths_range(deaths):
    if 0 <= deaths <= 49:
        return '0-49'
    elif 50 <= deaths <= 99:
        return '50-99'
    elif 100 <= deaths <= 199:
        return '100-199'
    elif 200 <= deaths <= 299:
        return '200-299'
    elif 300 <= deaths <= 399:
        return '300-399'
    elif 400 <= deaths <= 499:
        return '400-499'
    elif 500 <= deaths <= 999:
        return '500-999'
    elif 1000 <= deaths <= 1499:
        return '1000-1499'
    elif 1500 <= deaths <= 2000:
        return '1500-2000'
    elif 2001 <= deaths :
        return '> 2000'
    else:
        return 'Desconocido'

rangos_muertes = muertes_por_region.map(deaths_range)
rangos_muertes
```

```python
conteo_rangos = rangos_muertes.value_counts()

print("Conteo de regiones en cada rango de muertes:")
print(conteo_rangos)
```

```python
# Regiones con mas muertes dentro del rango '0-49'
muertes_primer_rango = muertes_por_region[rangos_muertes == '0-49']
muertes_primer_rango = muertes_primer_rango.sort_values(ascending=False)

print("Regiones con mas muertes (rango 0-49):")
muertes_primer_rango_top_10_paises = muertes_primer_rango.nlargest(10)
print(muertes_primer_rango_top_10_paises)
```

### Recuperaciones

```python
# Agrupar por region y calcular el numero de recuperaciones 
recuperaciones_por_region = df_current.groupby('region')['recovered'].sum()
recuperaciones_por_region
```

```python
region_max_recuperaciones = recuperaciones_por_region.idxmax()
max_recuperaciones = recuperaciones_por_region.max()
print(f"La región con más recuperaciones es: {region_max_recuperaciones},con {max_recuperaciones} recuperaciones.")
```

```python
top_10_paises_rec = recuperaciones_por_region.nlargest(10)

print("Los 10 países con más recuperaciones:")
print(top_10_paises_rec)
```

```python
# Aplicamos la funcion de map, para generar un rango de recuperaciones 
def recuperaciones_range(recuperaciones):
    if 0 <= recuperaciones <= 49:
        return '0-49'
    elif 50 <= recuperaciones <= 99:
        return '50-99'
    elif 100 <= recuperaciones <= 199:
        return '100-199'
    elif 200 <= recuperaciones <= 299:
        return '200-299'
    elif 300 <= recuperaciones <= 399:
        return '300-399'
    elif 400 <= recuperaciones <= 499:
        return '400-499'
    elif 500 <= recuperaciones <= 999:
        return '500-999'
    elif 1000 <= recuperaciones <= 1499:
        return '1000-1499'
    elif 1500 <= recuperaciones <= 2000:
        return '1500-2000'
    elif 2001 <= recuperaciones :
        return '> 2000'
    else:
        return 'Desconocido'

rangos_recuperaciones = recuperaciones_por_region.map(recuperaciones_range)
rangos_recuperaciones
```

```python
conteo_rangos_rec = rangos_recuperaciones.value_counts()

print("Conteo de regiones en cada rango de recuperaciones:")
print(conteo_rangos_rec)
```

```python
# Regiones con mas recuperaciones dentro del rango '0-49'
recuperaciones_primer_rango = recuperaciones_por_region[rangos_recuperaciones == '0-49']
recuperaciones_primer_rango = recuperaciones_primer_rango.sort_values(ascending=False)

print("Regiones con mas recuperaciones (rango 0-49):")
recuperaciones_primer_rango_top_10_paises = recuperaciones_primer_rango.nlargest(10)
print(recuperaciones_primer_rango_top_10_paises)
```

```python
df_con_muertes = df_current[df_current['deaths'] > 0]

# Calcular el epicentro (promedio ponderado por el número de muertes)
epicentro_lat = (df_con_muertes['lat'] * df_con_muertes['deaths']).sum() / df_con_muertes['deaths'].sum()
epicentro_long = (df_con_muertes['long'] * df_con_muertes['deaths']).sum() / df_con_muertes['deaths'].sum()

epicentro = (epicentro_lat, epicentro_long)

print(f"El epicentro del virus se encuentra en: {epicentro}")
```
